```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train=pd.read_csv("./train.csv")
```

```python
df_train
```

```python
df_y=df_train["claim"]
df_train.drop(columns="claim",inplace=True)
df_train.drop(columns="id",inplace=True)
df_y
```

```python
from sklearn.impute import SimpleImputer
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
df_train = imp_median.fit_transform(df_train)
```

```python
import sklearn
from sklearn.ensemble import RandomForestClassifier
import catboost
from catboost import CatBoostClassifier 
import lightgbm 
from lightgbm import LGBMClassifier
from sklearn.preprocessing import MinMaxScaler
import xgboost
from xgboost import XGBClassifier
from lightgbm import LGBMRegressor
```

```python
model=LGBMRegressor().fit(df_train,df_y)
```

```python
df_test=pd.read_csv("./test.csv")
df_test
```

```python
df_test.isnull().sum()
```

```python
df_test_id=df_test["id"]
df_test.drop(columns="id",inplace=True)
```

```python
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
df_test = imp_median.fit_transform(df_test)
```

```python
df_predict=model.predict(df_test)
```

```python
sub=pd.DataFrame({"id":df_test_id,"claim":df_predict})
sub.to_csv("submission1.csv",index=False)
```
